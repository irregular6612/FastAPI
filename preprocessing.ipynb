```python
import pandas as pd

# 전처리할 데이터 읽어오기
DB_path = "./datasets/GIST24-2.csv"
data = pd.read_csv(DB_path)

#타입 확인 및 최상단 데이터 샘플 불러오기
print(type(data))
data.head()
```

```python
# 불필요한 행 삭제
data = data.drop(['NO','세부분류','강/실/학','과정\n구분','강의실','강의\n계획서','강의언어','설명\n영상','이수구분','교과연구','수강\n정원'], axis=1)
```

```python
# 결측치 존재 열 결측치 채우기
data['담당교수'] = data['담당교수'].fillna("")
data['시간표'] = data['시간표'].fillna("")
```

```python
#확인
data.info()
```

```python
# 자료형 변환 확인 : pandas.DataFrame -> list of list(행 별로 분리된다.)
data.values
```

```python
# DB list 생성
DB = []

# list of list에서 요소(한 과목) 별 처리
for lec in data.values:
    
    # dict 형태로, fastapi 서버에서는 별도 클래스로 정의해서 사용.
    lecture = {}
    
    # 다른 처리가 필요 없는 자료는 바료 대입
    lecture['division'] = lec[0]
    lecture['code'] = lec[1]
    lecture['name'] = lec[2]
    lecture['professor'] = lec[3]
    
    # 시간표 자료는 요일 정보와 시간대 정보로 분리하여 리스트 2개로 저장.
    
    #요일
    day = [False, False, False, False, False]
    #시간대
    timePeriod = [False, False, False, False, False, False]
    
    if "월" in lec[4]:
        day[0] = True
    if "화" in lec[4]:
        day[1] = True
    if "수" in lec[4]:
        day[2] = True
    if "목" in lec[4]:
        day[3] = True
    if "금" in lec[4]:
        day[4] = True
    
    if "09:00~10:30" in lec[4]:
        timePeriod[0] = True
    if "10:30~12:00" in lec[4]:
        timePeriod[1] = True
    if "13:00~14:30" in lec[4]:
        timePeriod[3] = True
    if "14:30~16:00" in lec[4]:
        timePeriod[4] = True
    if "16:00~17:30" in lec[4]:
        timePeriod[5] = True
    
    # 시간 정보를 (요일 : list, 시간대 : list ) 형태로 저장.
    lecture['time'] = tuple((day, timePeriod))
    DB.append(lecture)
```
